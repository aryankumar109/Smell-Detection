```python
import os
import pandas as pd

# Specify the path to the folder containing the files
folder_path = '/content/drive/MyDrive/dense/5_Features/'

# Initialize an empty list to store DataFrames
dataframes_list = [[] for i in range(31)]
labels = []
V = []
A = []

emotion_labels_path = '/content/drive/MyDrive/dense/cleaned_data_readings.csv'
valence_label_path = '/content/drive/MyDrive/dense/cleaned_data_readings_valence.csv'
arousal_label_path = '/content/drive/MyDrive/dense/cleaned_data_readings_arousal.csv'
emotion_labels_data = pd.read_csv(valence_label_path)

# Iterate through each file in the folder
for index, row1 in emotion_labels_data.iterrows():
    file_name = row1['file_name']
    emotion_label = row1['Groups']
    print(emotion_label)
    val = row1['V']
    # aro = row1['A']

    # if file_name.endswith('.csv'):  # Adjust the file extension accordingly
    file_path = os.path.join(folder_path, f'{file_name}_eegfeatures1.csv')
    labels.append(emotion_label)
    V.append(val)
    # A.append(aro)

    # Read the file into a DataFrame
    df = pd.read_csv(file_path)  # Adjust the delimiter accordingly

    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        # Create a DataFrame for each row
        row_df = pd.DataFrame(row).T

        # Append the row DataFrame to the list
        dataframes_list[index].append(row_df)

# Concatenate all DataFrames in the list into a single DataFrame
dataframes_file_v = []
for i in range(31):
    final_dataframe = pd.concat(dataframes_list[i], ignore_index=True)
    dataframes_file_v.append(final_dataframe)

# Display the resulting DataFrame
print(len(dataframes_file_v))
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
from scipy.stats import ttest_rel
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
# Assuming you have a dataset with features (X) and labels (y)
# Replace X_train, y_train, X_test, y_test with your actual data
# Make sure your data is appropriately preprocessed and formatted
label_encoder = LabelEncoder()
Y_v = label_encoder.fit_transform(Y_v)
# Y_a = label_encoder.fit_transform(Y_a)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from scipy import stats
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
results_df = pd.DataFrame(columns=['Accuracy'])
for i in range(31):
    X_train, X_test, y_train_v, y_test_v = train_test_split(dataframes_file_v[i], Y_v, test_size=0.2,random_state=21,shuffle=True)
    acc_val = []
    # X_train = X_train.to_numpy()
    # X_test = X_test.to_numpy()
    for i in range(1,100):
        knn = KNeighborsClassifier(n_neighbors=i)

        # Train the classifier
        knn.fit(X_train, y_train_v)

        # Make predictions on the test set
        y_pred = knn.predict(X_test)

        accuracy_a = accuracy_score(y_test_v, y_pred)
        acc_val.append(accuracy_a)

    acc_val = np.asarray(acc_val)
    index = acc_val.argmax()

    knn = KNeighborsClassifier(n_neighbors=index+1)

    # Train the classifier
    knn.fit(X_train, y_train_v)

    # Make predictions on the test set
    y_pred = knn.predict(X_test)

    # Calculate the accuracy
    accuracy_v = accuracy_score(y_test_v, y_pred)
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test_v, y_pred, average='binary')
    print(y_test_v, y_pred)
    print(f"Accuracy(HVLV): {accuracy_v}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1_score}")
    print(y_test_v, y_pred)

    # Print classification report
    print("Classification Report:\n", classification_report(y_test_v, y_pred))

    # new_row = {'Column1': 1, 'Column2': 'A', 'Column3': 10.5}
    # empty_df = empty_df.append(new_row, ignore_index=True)
    results_df = results_df.append({'Accuracy': accuracy_v, 'Precision': precision, 'Recall': recall, 'F1 Score': f1_score},ignore_index = True)
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import pandas as pd
import numpy as np

results_df_a = pd.DataFrame(columns=['Accuracy', 'Precision', 'Recall', 'F1 Score'])

num_features = 25  # Specify the number of input features

for i in range(31):
    X_train, X_test, y_train_a, y_test_a = train_test_split(dataframes_file_v[i].iloc[:, :num_features], Y_v, test_size=0.2, random_state=21, shuffle=True)

    param_dist = {
        'hidden_layer_sizes': [(50,), (100,), (150,)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'max_iter': [100, 200, 300]
    }

    mlp = MLPClassifier(random_state=42)
    randomized_search = RandomizedSearchCV(estimator=mlp, param_distributions=param_dist, n_iter=10, cv=3, n_jobs=-1, verbose=1, random_state=42)

    randomized_search.fit(X_train, y_train_a)

    best_params = randomized_search.best_params_

    mlp = MLPClassifier(random_state=42, **best_params)

    # Train the classifier
    mlp.fit(X_train, y_train_a)

    # Make predictions on the test set
    y_pred = mlp.predict(X_test)

    # Calculate the accuracy
    accuracy_a = accuracy_score(y_test_a, y_pred)

    # Calculate precision, recall, and F1-score
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test_a, y_pred, average='binary')

    print(y_test_a, y_pred)
    print(f"Accuracy(HALA): {accuracy_a}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1_score}")

    # Print classification report
    print("Classification Report:\n", classification_report(y_test_a, y_pred))

    # Append the results to the results dataframe
    results_df_a = results_df_a.append({'Accuracy': accuracy_a, 'Precision': precision, 'Recall': recall, 'F1 Score': f1_score}, ignore_index=True)
```
